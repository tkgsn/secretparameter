```python
import pickle
n_univ = 10000
n_iter = 10000
with open(f"data/univ_{n_univ}_para_{0.5}_n_iter_{n_iter}_temp_.pkl", "rb") as f:
    results_public, results_secret = pickle.load(f)
#with open(f"data/univ_{n_univ}_data_{0.5}_n_iter_{n_iter}_para_.pkl", "rb") as f:
#    results_public, results_secret = pickle.load(f)
```

```python
import japanize_matplotlib
import matplotlib.pyplot as plt
data_ratios = list(results_public.keys())
plt.plot(data_ratios, [result[0][0] for _, result in results_public.items()], linestyle="dashed", marker="o", label="public MLE")
plt.plot(data_ratios, [result[1][0] for _, result in results_public.items()], linestyle="dashed", marker="x", label="public unbiased estimator")
plt.plot(data_ratios, [result[0][0] for _, result in results_secret.items()], marker="s", label="secret MLE")
plt.plot(data_ratios, [result[1][0] for _, result in results_secret.items()], marker="+", label="secret unbiased estimator")
plt.legend(fontsize=10, loc='center',bbox_to_anchor=(1/2, 3.23/5))
plt.xlabel("$\pi_{d_0}$", fontsize=15)
plt.ylabel("Average of the error", fontsize=15)
plt.xlim(0, 1)
#plt.xlim(0.4489,0.6530)

plt.savefig(f"results/univ_{n_univ}_data_{0.5}_n_iter_{n_iter}_para.eps",bbox_inches="tight", pad_inches=0.0)
```
